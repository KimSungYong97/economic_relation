```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
countries = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))

countries.tail(3)
```

```python
ax = countries.plot(column="continent", legend=True, categorical=True)
ax.set_title(" ")
ax.set_axis_off()
plt.show()
```

```python
import pandas as pd
country=pd.read_csv('./data/geo_country.csv')
```

```python
countries=countries.sort_values(by='pop_est')
country=country.sort_values(by='pop_est')
countries=countries.reset_index(drop=True)
country=country.reset_index(drop=True)
country
```

```python
countries['cluster_id']=country['cluster_id'].copy()
countries['color']=country['color'].copy()
```

```python
colors=['gray','Reds','Blues','Greens']
```

```python
indexNames = countries[ (countries['continent'] == 'Antarctica') ].index
countries.drop(indexNames , inplace=True)
```

```python
countries
```

```python
ax = countries.plot(column="cluster_id", legend=True, categorical=True,color=countries['color'],edgecolor="k")
ax.set_title("cluster_map")
ax.set_axis_off()
plt.rcParams["figure.figsize"] = (20,18)
plt.show()
```
